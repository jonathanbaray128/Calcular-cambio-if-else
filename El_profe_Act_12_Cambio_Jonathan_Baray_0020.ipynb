```python
# Calculando el cambio usando if else
# datos de entrada CD, CP
# mostrar Cambio si es falso
# si es verdadero mostrar
CD = float(input("Ingrese la ingrese el costo del producto: "))
CP = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
   print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
  print("Fin del programa")
  print("Programa realizado por Jonathan Baray NL = 06")
```
